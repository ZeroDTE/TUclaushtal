```python
# D3 Kelvin in Fahrenheit umrechnen

def kel_in_fahr():
    temps = []
    for i in range(5):
        kelvin = float(input("Temperature in Kelvin eingeben oder 00 zum Abbrechen"))

        if kelvin == 00 : 
            break
        else: 
            fahrenheit = ((kelvin - 273.15) * 1.8) + 32
            fahrenheit = round(fahrenheit,2)
            temps.append(fahrenheit)
    
    return temps

    
    
```

```python
# d6 

def dez_bin(zahl):

    if zahl == 0 : 

        return "0"
    
    number = abs(zahl)
        
    binary = ""
    while number > 0 :
        rest = number%2 
        binary = str(rest)  + binary
        number = number // 2


    return binary
```
