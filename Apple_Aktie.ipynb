```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn.preprocessing import MinMaxScaler
from matplotlib.animation import  FuncAnimation, PillowWriter,writers
```

```python
# Laden der CSV-Datei mit Angabe des Trennzeichens
file_path = 'Apple_Dataset.csv'
df = pd.read_csv(file_path, delimiter=',')

df.info()
#df.describe()
```

```python
# Anzeigen der ersten Zeilen des DataFrame
df.head()
df['Date'] = pd.to_datetime(df['Date'])
# Daten nach 'High' sortieren
df_sorted = df.sort_values(by='Date', ascending = False)
df_sorted.head()
```

```python
# Löschung von Spalten

df2= df.drop(['Low', 'Open', 'High', 'Adj Close'], axis=1)
df2
```

```python

# Daten und Schluss Atkienpreise über die Jahre plotten
filtered_df = df2[df2['Date'] > '2010-01-01']


plt.plot(filtered_df['Date'], filtered_df['Close'], 'g', label='Schlusskurs', linewidth = 2)

plt.legend()
plt.xlabel("Jahre")
plt.yscale('log')


plt.show()
```

```python
# Daten und Schluss Atkienpreise über die Jahre plotten.(Andere Methode)
df.plot.line(y="Close", use_index=True)

df.plot.line(y="Close", use_index=True, logy=True)
```

```python
x = np.arange(len(filtered_df))
y = filtered_df["Close"]


fig, ax = plt.subplots()
line, = ax.plot(x, y, color='k')

def update(frame):
    ax.clear()
    ax.plot(x[:frame], y[:frame], color='k')
    ax.set_xlim(0, len(x))
    ax.set_ylim(min(y) - 1, max(y) + 1)
    ax.set_title(f"Frame {frame}")

# Create animation
ani = FuncAnimation(fig, update, frames=len(x), repeat=False)

# Save the animation as a GIF file
writer = PillowWriter(fps=10)
ani.save('animation.gif', writer=writer)

# Display the animation
plt.show()
```

```python
# zweiter plot (Volumen)

plt.bar(filtered_df['Date'] , filtered_df['Volume'], color='b', label='Volume')
plt.xlabel('Date')
plt.ylabel('Volumen')
plt.title('Aktien-Volumen über die Jahre')
plt.legend()
plt.grid(True)
plt.show()
```

# Linear Regression

```python
from scipy.stats import linregress
```

```python
x = np.arange(len(filtered_df))
y = filtered_df["Close"]


# Durchführung der linearen Regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Regressionslinie berechnen
filtered_df['Regressionslinie'] = intercept + slope * x 
y_dach=filtered_df['Regressionslinie']

# Ergebnis anzeigen
print(filtered_df.head(-10))

# Optional: Plotten der Daten und der Regressionslinie
import matplotlib.pyplot as plt

plt.plot(x,y, label='Daten')
plt.plot(x,y_dach, color='red', label='Regressionslinie')
plt.legend()
plt.show()
```

```python
# Genauigkeitsmetriken berechnen
from sklearn.metrics import mean_absolute_error, r2_score
mae = mean_absolute_error(y, y_dach)
r2 = r2_score(y, y_dach)

print(f'Mean Absolute Error (MAE): {mae}')
print(f'R^2 Score: {r2}')
```
