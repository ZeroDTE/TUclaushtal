```python
def cels_kel(eingabe):
    kel = eingabe -  absl_nl_p
    return kel

def cels_fahr(eingabe): #F = (°C × 9/5) + 32
    fahrenheit = (eingabe * 9/5) + 32
    return fahrenheit

def kel_cels(eingabe) : 
    celsius = eingabe - absl_nl_p

    return celsius


def kel_fahr(eingabe) : 
    fahreheit = 1.8 * (eingabe + absl_nl_p) + 32

    return fahreheit


def fahr_cels(eingabe) : 
    
    celsuis = (eingabe * 5/9) - 32

    return celsuis


def fahr_kel(eingabe) : 
    fahrenheit_cels = fahr_cels(eingabe)
    kelvin = cels_kel(fahrenheit_cels)

    return kelvin

```

```python
while True : 
    print("(1) Umrechnung von Celsius nach Kelvin")
    print("(2) Umrechnung von Celsius nach Fahrenheit")
    print("(3) Umrechnung von Kelvin nach Celsius")
    print("(4) Umrechnung von Kelvin nach Fahrenheit")
    print("(5) Umrechnung von Fahrenheit nach Celsius")
    print("(6) Umrechnung von Fahrenheit nach Kelvin")
    print("(0) Programm Beenden")
    
    choice = input("wähle eine Option aus der Liste aus")
    
    if choice == "0":
        print("Programm wird beendet")
        break
    
    
    temp = float(input("Bitte Temperature eingeben"))
    
    if choice == "1" : 
        result = cels_kel(temp)
        print(f"{temp} Celcuis ist {result} in kelvin")

    elif choice =="2": 
        result = cels_fahr(temp)
        print(f"{temp} Celcuis ist {result} in Fahrenheit")

    elif choice =="3": 
        result = kel_cels(temp)
        print(f"{temp} Kelvin ist {result} in Celsius")

    elif choice =="4": 
        result = kel_cels(temp)
        print(f"{temp} Kelvin ist {result} in Celsius")


    elif choice =="5": 
        result = fahr_kel(temp)
        print(f"{temp} Fahrenheit ist {result} in Celsius")


    elif choice =="6": 
        result = kel_cels(temp)
        print(f"{temp} Kelvin ist {result} in Celcuis")

```
