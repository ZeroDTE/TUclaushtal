# Linear Regression

```python
from scipy.stats import linregress
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

```

```python
# x enthält ein Array von 0 bis zur Länge des gefilterten Dataframes (minus 1)
# z.B. wenn filtered_df 100 Zeilen hat, dann enthält x: [0,1,2,...,98,99]
# Dies wird als x-Achse für die Regression verwendet, wobei jeder Datenpunkt einen Index darstellt
x = np.arange(len(filtered_df))
# Hole die Schlusskurse als y-Werte
y = filtered_df["Close"]


# Durchführung der linearen Regression
# slope = Steigung, intercept = y-Achsenabschnitt
# r_value = Korrelationskoeffizient, p_value = Signifikanzwert
# std_err = Standardfehler der Steigung
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Berechne die Regressionslinie mit der Formel y = mx + b
# wobei m = slope (Steigung) und b = intercept (y-Achsenabschnitt)
filtered_df['Regressionslinie'] = intercept + slope * x 
# Speichere die berechneten y-Werte in y_dach (y mit Dach = geschätzte y-Werte)
y_dach=filtered_df['Regressionslinie']

# Zeige die ersten Zeilen des Dataframes an (ohne die letzten 10 Zeilen)
print( filtered_df.head(-10))



# Plotte die originalen Datenpunkte
plt.plot(x,y, label='Daten')
# Plotte die Regressionslinie in rot
plt.plot(x,y_dach, color='red', label='Regressionslinie')
# Füge eine Legende hinzu
plt.legend()
# Zeige den Plot an
plt.show()
```

```python
# Genauigkeitsmetriken berechnen
from sklearn.metrics import mean_absolute_error, r2_score
mae = mean_absolute_error(y, y_dach)
r2 = r2_score(y, y_dach)

print(f'Mean Absolute Error (MAE): {mae}')
print(f'R^2 Score: {r2}')
```

# Polynomiale Linear Regression

```python
# Erstelle eine größere Abbildung (12x8 Zoll) für bessere Lesbarkeit
plt.figure(figsize=(12,8))

# Linear Regression (wie oben bereits durchgeführt)
# x-Werte sind die Tagesindizes (0,1,2,...), y-Werte sind die Schlusskurse
x = np.arange(len(filtered_df))  # z.B. [0,1,2,3,...]
y = filtered_df["Close"]         # z.B. [190.5, 191.2, 189.8,...]
# Berechne Parameter der linearen Regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)
# Berechne die vorhergesagten y-Werte mit y = mx + b 
y_linear = intercept + slope * x # z.B. wenn slope=0.5 und intercept=100: y = 0.5*x + 100

# Polynomiale Regression
# Grad 3 bedeutet eine Funktion der Form: y = ax³ + bx² + cx + d
degree = 5
# Erstelle Pipeline für polynomiale Regression:
# 1. Transformiere x in polynomiale Features (x, x², x³)
# 2. Führe lineare Regression auf diesen Features durch
# z.B. x=[1,2,3] wird zu x.reshape(-1,1)=[[1],[2],[3]] (2D Array mit einer Spalte)
poly_model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
poly_model.fit(x.reshape(-1, 1), y)
# Berechne vorhergesagte Werte für polynomiales Modell
y_poly = poly_model.predict(x.reshape(-1, 1))

# Visualisierung der Daten und Modelle
# Streue die tatsächlichen Datenpunkte
plt.scatter(x, y, label='Daten', color='blue')
# Zeichne die lineare Regressionslinie
plt.plot(x, y_linear, color='red', label=f'Lineare Regression (R² = {r_value**2:.4f})')
# Zeichne die polynomiale Regressionslinie
plt.plot(x, y_poly, color='green', label=f'Polynomiale Regression (Grad={degree})')

# Beschriftungen hinzufügen
plt.title('Aktienkurs: Lineare vs Polynomiale Regression')
plt.xlabel('Tage Index')
plt.ylabel('Aktienkurs')
plt.legend()

# Berechne Genauigkeitsmetriken für beide Modelle
# MAE = Durchschnittliche absolute Abweichung zwischen Vorhersage und tatsächlichem Wert
# R² = Bestimmtheitsmaß (1 = perfekte Vorhersage, 0 = schlechte Vorhersage)
mae_linear = mean_absolute_error(y, y_linear)
r2_linear = r2_score(y, y_linear)
mae_poly = mean_absolute_error(y, y_poly)
r2_poly = r2_score(y, y_poly)

print(f'Lineare Regression - MAE: {mae_linear:.4f}, R² Score: {r2_linear:.4f}')
print(f'Polynomiale Regression - MAE: {mae_poly:.4f}, R² Score: {r2_poly:.4f}')

# Vorhersage der nächsten 5 Tage
# Erstelle neue x-Werte für die Zukunft (nächste 5 Tage)
future_x = np.arange(len(filtered_df), len(filtered_df) + 5)
# Berechne Vorhersagen mit beiden Modellen
future_y_linear = intercept + slope * future_x  # Lineare Vorhersage
future_y_poly = poly_model.predict(future_x.reshape(-1, 1))  # Polynomiale Vorhersage

# Zeichne die Vorhersagen als gestrichelte Linien
plt.plot(future_x, future_y_linear, color='red', linestyle='--', label='Lineare Vorhersage')
plt.plot(future_x, future_y_poly, color='green', linestyle='--', label='Polynomiale Vorhersage')

plt.legend()
plt.show()

# Gib die konkreten Vorhersagewerte aus
print("\nVorhersage der nächsten fünf Tage:")
for i, (linear_pred, poly_pred) in enumerate(zip(future_y_linear, future_y_poly), 1):
    print(f"Tag +{i}: Linear: {linear_pred:.2f}, Polynomial: {poly_pred:.2f}")
```
