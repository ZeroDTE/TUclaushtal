```python
#2) While Schleife 
antwort=input("Möchtest du Feieraben haben, beantworte die Frage mit J/N").capitalize()
wiederholungen = 1

while antwort not in ["J","N"]:
    print("Eingabe ungültig")
    antwort=input("Möchtest du Feieraben haben, beantworte die Frage mit J/N").capitalize()
    wiederholungen += 1



print(f"Vielen Dank, du hast {wiederholungen}-mal gebraucht")


    
```

```python
# Fakultät 

#fak(5) = 5 * 4 * 3 *2 * 1

#Iterativ
def fak(n):
    fakulaet = 1
    for n in range(1,n+1):
        fakulaet = fakulaet * n

    return fakulaet
    
#rekursiv
def fak_rekursive(n): 
    if n  == 0 : 
        return 1 

    else :
        return n * fak_rekursive(n-1)
```

```python
fak(5),fak_rekursive(5)
```

```python
fak_rekursive(5)
# 5 * 4 * 3  * 2
```
